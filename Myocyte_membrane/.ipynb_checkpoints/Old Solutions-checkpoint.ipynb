### Exercise: A simple voltage clamp model

#### Exercise 1a)

The first thing we do is to define the function
$$V_{\rm target} = \begin{cases}
-40 {\rm\ mV} & \mbox{if } 2 {\rm\ ms} < t < 6 {\rm\ ms}, \\
-80 {\rm\ mV}& \mbox{else.}
\end{cases}$$
We do this by using boolean math: `(2 < t)*(t < 6)` instead of an if-test, this way the function can operator on a whole time array as input

```python
def V_target(t):
    return -80 + 40*(2 < t)*(t < 6)
```

We can check that we have written the function correctly by plotting it

```python
t = np.linspace(0, 10, 1001)
plt.plot(t, V_target(t))
plt.show()
```

**Exercise 1b)**

Our ODE is as follows:
$$\frac{{\rm d}V}{{\rm d}t} = \frac{V_{\rm target} - V}{C_{\rm m} R_{\rm s}}.$$
We want to analyze the units and check that they are consistent.

If want to measure $V$ in mV and time in ms. This means the left hand side is given in units of ,V/ms (or equivalently V/s). The right hand side is given in units of mV divided by nF times M$\Omega$, nF times M$\Omega$ works out to be ms, so the RHS is also mV/ms.

So we see that giving the capacitance of the membrane in nF and the resistance of the pipette in M$\Omega$ is consistent.

**Exercise 1c)**

Implement the ODE and use `odeint` to solve it to find the membrane potential $V(t)$ in the period $t \in [0, 10]$ ms. Let $C_m = 0.05$ nF and $R_s = 10$ M$\Omega$. Let $V$ be denoted in mV and time in ms. Plot your solution.

```python
def rhs(V, t, Cm, Rs):
    return (V_target(t) - V)/(Cm*Rs)

Cm = 0.05
Rs = 10
params = (Cm, Rs)

dt = 0.01
T = 10
time = np.arange(0, T, dt)

V0 = -80

V = odeint(rhs, V0, time, params)
```

```python
plt.plot(time, V)
plt.xlabel('Time [ms]')
plt.ylabel('Membrane Potential [mV]')
plt.show()
```

To make this easier, we plot them overlapping

```python
plt.plot(time, V)
plt.plot(time, V_target(time))
plt.xlabel('Time [ms]')
plt.ylabel('Membrane Potential [mV]')
plt.legend(["V", "V_target"])
plt.show()
```


**(Optional) Exercise 1e)**

Solve the ODE by hand. What kind of ODE/solution is this? What is the most important functional parameter in this solution?

The ODE is:
$$\frac{{\rm d}V}{{\rm d}t} = \frac{V_{\rm target} - V}{C_{\rm m} R_{\rm s}}.$$
If we redefine $u = V_{\rm target} - V$, we see that the ODE is separabel:
$$-\frac{{\rm d}u}{{\rm d}t} = \frac{u}{C_{\rm m} R_{\rm s}}.$$
So we flip it around to be
$$- \frac{1}{u} {\rm\ d}u = \frac{1}{C_{\rm m} R_{\rm s}} {\rm\ d}t.$$
Integrating gives
$$\ln u = -\frac{t}{C_{\rm m} R_{\rm s}} + C.$$
or
$$u = Ae^{-t/(C_{\rm m}\cdot R_{\rm s})}.$$
Now we insert for $u$ so we get
$$V(t) = V_{\rm target} - Ae^{-t/(C_{\rm m}\cdot R_{\rm s})}.$$
We know that $V(0) = V_0$ so we get
$$V(t) = V_{\rm target} - (V_{\rm target} - V_0)e^{-t/(C_{\rm m}\cdot R_{\rm s})}.$$
This means the function is an exponential decay towards the target value, with a time constant of $\tau = C_{\rm m}\cdot R_{\rm s}.$

### Exercise 2: Calculating Nernst Potentials

The following table gives typical values for mammalian cardiomyocytes, and is taken from [Katz, 2010]

| Ion        | Intracellular | Extracellular  |
| -----------|:-------------:| -----:|
| Sodium     | 8 mM          | 110 mM |
| Potassium  | 100 mM        | 4 mM  |
| Calcium    | 0.2 µM        | 1 mM  |

You will also need to know these physical constants

| Name | Symbol  | Value | Unit |
| -----------|:-------------:|-----:|------:|
| Faraday's constant  | $F$ | 96485 | C/mol |
| Gas Constant | $R$ |  8.314 | J/(mol K) |
| Temperature | $T$ | 310|  K | 


**Exercise 2a)**

Based on the tabulated values, calculate the Nernst potentials for the three ionic species. 

All the physical parameters are in SI units, so our answers will be in Volts, so we multiply our answers by 1000 to get mV instead.

```python
F = 96485
R = 8.314
T = 310

E_Na = R*T/F * np.log(110/8)
E_K = R*T/F * np.log(4/100)
E_Ca = R*T/(2*F) * np.log(1/0.0002)

print("E_Na = {:.1f} mV".format(E_Na*1e3))
print("E_K = {:.1f} mV".format(E_K*1e3))
print("E_Ca = {:.1f} mV".format(E_Ca*1e3))
```

**Exercise 2b)**

You have received two positive answers, and one negative. Explain why this is the case.

Looking at the concentrations of the three ions. We see that there is a large inward concentration gradient for sodium and calcium, which is why their Nernst potential is positive, because a higher potential inside than outside will balance out the inward concentration gradient. For potassium however, the concentration gradient is the other direction, so we need to have a potential difference in the other direction to balance this gradient, which is why we get a negative potential in that case.


**Exercise 2c)**

Assume we have a membrane that is at a potential of -80 mV. We now open sodium channels in the membrane. Sodium ions will now start to enter the cell, increasing the membrane potential untill it reaches the Nernst potential for sodium in the cell. Estimate the number of Na$^+$-ions that cross the membrane for this change to occur
* *Hint 1:* Recall that $\Delta Q = C_m \Delta V$.
* *Hint 2:* You can use Faraday's constant to go from a charge to the number of charged particles.


First, we have the change in potential. We start at -80 mV and go to $E_{\rm Na}$, which we found to be 70 mV. So we have 
$$\Delta V = 150 {\rm\ mV}.$$
Multiplying with $C_{\rm m} = 0.05 {\rm\ nF}$, we get the total charge buildup
$$\Delta Q = 0.05 {\rm\ nF} \cdot 150 {\rm\ mV} = 7.5 {\rm\ pC}.$$

Next, to go from coloumbs to the number of ions, we use Faraday's constant:
$$\frac{7.5 {\rm\ pC}}{96485 {\rm\ C/mol}} = 7.8\cdot 10^{-17} {\rm\ mol}.$$
The number of ions moving across the membrane is then given by
$$7.8\cdot 10^{-17} {\rm\ mol} \cdot 6.022\cdot10^{23} {\rm\ ions/mol} = 4.7 \cdot 10^7 {\rm\ ions}.$$
This might sound like a lot, but it is actually a fairly small amount of ions.

Assuming the cell has a total volume of 20 picoliters, how much does the sodium concentration change as a result of this?

The tabulated concentration is 110 mM. We found that $7.8\cdot10^{-17}$ mol, so if we assume a volume of 20 picoliters, we find a concentration change of
$$\frac{\rm d[Na^{+}]}{{\rm d} t} = \frac{7.8\cdot10^{-17} {\rm\ mol}}{20 {\rm\ pL}} = 3.9 {\rm\ µM}.$$
So to go from -80 mV to 70 mV due to a sodium current, there will only be a change in the sodium concentration of a few µM, compared to the original ~100 mM, this is a relative change on the order of $10^{-5}$. 

**Exercise 2d)**

Assume the intracellular sodium increases by 30%, how much does the Nernst potential change?

We have seen that a change of 30% is not realistic, but let us see what happens to the Nernst potential if we assume this to be the case.

```python
F = 96485
R = 8.314
T = 310

original = R*T/F * np.log(110/8)
reduction = R*T/F * np.log(110*0.7/8) # Reduce by 30%
increase = R*T/F * np.log(110*1.3/8) # Increase by 30%

print("Original:        {:.1f} mV".format(original*1e3))
print("Reduced by 30%:  {:.1f} mV".format(reduction*1e3))
print("Increased by 30% {:.1f} mV".format(increase*1e3))
```

We see that even a shift as dramatic as 30% shifts the Nernst potential by only 10 mV. Considering the differences between the different species' Nernst potentials are 40-150 mV, a change of 10 mV is relatively minute.


**Exercise 2e)**

Based on your answers to the last two exercises. Explain why the Nernst potential can be considered a constant, despite it being a function of the intracellular and extracellular concentrations.

In 2c) we saw that the relative change in sodium concentration is on the order of $10^{-5}$, and in 2d) we saw that a much more dramatic change than this only shifted the Nernst potential by 10 mV, so the actual shift during an action potential is probably much less than 1 mV, meaning we can consider it constant.

**Note:** The Nernst potential is usually considered a constant for most species, but due to the initially very low intracellular calcium, the Nernst potential of calcium changes more drastically than for other species, which potentially can be a source of error in models.

### Exercise 3: The Equilibrium Membrane Potential

**Exercise 3a)**
Solve the following ODE numerically
$$C_{\rm m}\frac{{\rm d}V}{{\rm d}t} = - g_{\rm Na} (V-E_{\rm Na}).$$

Let $C_{\rm m} = 0.05$ nF and $g_{\rm Na}=0.005$  µS. Let the membrane potential be given in units of mV and time in units if ms. Make sure all units are compatible, and make any necessary changes.

Assume an initial membrane potential of $V_0 = -80$ mV, and solve the ODE for $t\in[0, 200]$ ms.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

Cm = 0.05 # nF
g_Na = 0.005 # µS
E_Na = 70 # mV

def rhs(V, t):
    dV_dt = -g_Na*(V - E_Na)/Cm
    return dV_dt

T = 200
dt = 0.01
time = np.arange(0, T, dt)
V0 = -80

V = odeint(rhs, V0, time)
```

```python
plt.plot(time, V)
plt.xlabel('Time [ms]')
plt.ylabel('Membrane Potential [mV]')
plt.show()
```

**(Optional) Exercise 3b)**

Solve the ODE by hand. What is the time constant in the exponential relaxation?

This ODE is almost identical to what we solved earlier, and we carry out the same process. First we substitute $u = V-E_{\rm Na}$ to find:
$$\frac{1}{u} \frac{{\rm d}u}{{\rm d}t} = -\frac{g_{\rm Na}}{C_{\rm m}}.$$
Which is a separabel ODE, so we get
$$u(t) = Ce^{-t\cdot g_{\rm Na}/C_{\rm m}}.$$
So we get
$$V(t) = E_{\rm Na} + (V_0 - E_{\rm Na})e^{-t/\tau},$$
where $\tau = g_{\rm Na}/C_{\rm m}$. So we see the solution is an exponential decay to the Nernst potential, with a time constant of $g_{\rm Na}/C_{\rm m}$. Thus, the bigger the conductance of the channels, the bigger the current, and the faster the change in potential, and the bigger the capacitance, the harder it is to change the membrane potential, so it takes longer.


**Exercise 3c)**

Plot your solution. How long does it take for the membrane potential to equilibriate? Compare this with your time constant from 3b (if you did it).

```python
plt.plot(time, V)
plt.xlabel('Time [ms]')
plt.ylabel('Membrane Potential [mV]')
plt.axvline(10, color='black')
plt.show()
```

We see it takes about 10 ms before the difference to the equilbrium value has decayed to $1/e$'th, so after ~30 ms we are pretty much at equilibrium

The time constant we found analytically is
$$\tau = \frac{g_{\rm Na}}{C_{\rm m}} = \frac{0.005 {\rm\ µS}}{0.05 {\rm\ nF}} = 10 {\rm\ ms}.$$
Which seems to be right from the plot.

** Exercise 3d)**

Now add a calcium current and a potassium current to your ODE system. Let $g_{\rm Ca} = 0.002$ µS and $g_{\rm K} = 0.05$ µS. Solve the equation from an initial potential of $V_0 = 0$ mV for $t \in [0, 200]$ ms.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

Cm = 0.05 # nF
g_Na = 0.005 # µS
E_Na = 70 # mV
g_K = 0.05 # µS
E_K = -85 # mV
g_Ca = 0.002 # µS
E_Ca = 110 # mV

def rhs(V, t):
    dV_dt = -(g_Na*(V - E_Na) + g_K*(V - E_K) + g_Ca*(V - E_Ca))/Cm
    return dV_dt

T = 200
dt = 0.01
time = np.arange(0, T, dt)
V0 = 0

V = odeint(rhs, V0, time)
```

**Exercise 3e)**

Plot your solution. What is the equilibrium potential of the system now? Explain why. Does the initial membrane potential impact what the equilibrium potential is?

```python
plt.plot(time, V)
plt.xlabel('Time [ms]')
plt.ylabel('Membrane Potential [mV]')
plt.show()
```

The equilibrium potential is about -65. We can write it out by printing out `V[-1]`, as `[-1]` means the last element of an array:

```python
print(V[-1])
```

We can also rerun and change $V_0$, the equilibrium potential is not affected by the initial value.
